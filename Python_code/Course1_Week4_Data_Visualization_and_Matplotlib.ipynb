```python
import json
import time
path = "/Users/yaohanli/Downloads/yelp_dataset/review.json"
f = open(path, 'r')
```

```python
dataset = []
for i in range(50000):
    dataset.append(json.loads(f.readline()))
```

```python
datasetWithTimeValues = []
```

```python
for d in dataset:
    d['date']
    d['timeStruct'] = time.strptime(d['date'], "%Y-%m-%d %H:%M:%S")
    d['timeInt'] = time.mktime(d['timeStruct'])
    datasetWithTimeValues.append(d)
```

```python
from collections import defaultdict
```

```python
weekRatings = defaultdict(list)
```

```python
for d in datasetWithTimeValues:
    day = d['timeStruct'].tm_wday
    weekRatings[day].append(d['stars'])
```

```python
weekAverages = {}
```

```python
for d in weekRatings:
    weekAverages[d] = sum(weekRatings[d]) * 1.0 / len(weekRatings[d])
```

```python
weekAverages
sorted_dict = sorted(weekAverages.items(), key = lambda t: t[0])
# sort the list
```

```python
X = list(sorted_dict)
```

```python
Y = [sorted_dict[x] for x in X]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X, Y)
```

```python
plt.bar(X, Y)
```

```python
plt.ylim(3.6, 3.8)
plt.bar(X, Y)
```

```python
plt.ylim(3.6, 3.8)
plt.xlabel("Weekday")
plt.ylabel("Rating")
plt.xticks([0, 1, 2, 3, 4, 5, 6], ['S', 'M', 'T', 'W', 'TH', 'F', 'S'])
plt.title("Rating as a function of weekday")
plt.bar(X, Y)
```
